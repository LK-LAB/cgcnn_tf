```python
import tensorflow as tf
from cgcnn.data_tf import CIFData, collate_pool, Dataloader, CIFData_from_DataFrame
from cgcnn.model_tf import CrystalGraphConvNet
import pandas as pd
```

```python
import functools
import os
import json
import numpy as np
from pymatgen.core.structure import Structure
import math
import warnings
```

```python
# dataset = CIFData("data/sample-regression/")
df = pd.read_excel("Test-file.xlsx")
df.columns = ["id", "target", "cif"]
dataset = CIFData_from_DataFrame(df.iloc[:100])
```

```python
structures, _, _ = dataset[0]
orig_atom_fea_len = structures[0].shape[-1]
nbr_fea_len = structures[1].shape[-1]
```

```python
cgcnn_m = CrystalGraphConvNet(orig_atom_fea_len, nbr_fea_len,
                            atom_fea_len=64,
                            n_conv=3,
                            h_fea_len=128,
                            n_h=1)
cgcnn_m.compile(optimizer='adam', loss='mean_squared_error',metrics=[tf.keras.metrics.RootMeanSquaredError(), 'mae'], run_eagerly=True)
```

```python
batched = Dataloader(dataset, batch_size=1)
```

```python
history = cgcnn_m.fit(batched, epochs=1)
```
